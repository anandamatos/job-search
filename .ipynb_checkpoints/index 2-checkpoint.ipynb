
---

#### **3. Identifica√ß√£o das Entidades**  
- **Objetivo**: Listar objetos principais que armazenar√£o dados.  
- **Classifica√ß√£o**:  
  - **Entidades Fortes**: Existem independentemente (ex.: "Confer√™ncia").  
  - **Entidades Fracas**: Dependem de outras (ex.: "Apresenta√ß√£o" depende de "Sess√£o").  
- **Exemplo**:  
  - Confer√™ncia, Participante, Sess√£o, Artigo, Revisor.  

### **3.1 Identifica√ß√£o das Entidades para Gest√£o Inteligente de Vagas**  
*(Com Foco em Automa√ß√£o e Rastreamento de Processos)*  

---

#### **Entidades Fortes (Independentes)**  
| Entidade          | Descri√ß√£o                                                                 | Atributos Chave (Exemplos)                                                                 |  
|-------------------|---------------------------------------------------------------------------|-------------------------------------------------------------------------------------------|  
| **Candidato**     | Representa voc√™ (usu√°rio do sistema)                                      | `id`, `habilidades_tecnicas`, `pretensao_salarial`, `github_curriculo`, `batch_atual`     |  
| **Empresa**       | Organiza√ß√µes que publicam vagas                                           | `id`, `nome`, `setor`, `porte`, `reputacao_glassdoor`                                     |  
| **Plataforma**    | Fonte das vagas (LinkedIn, Glassdoor, etc.)                               | `id`, `nome`, `url_base`, `dificuldade_scraping` (n√≠vel de complexidade para extra√ß√£o)     |  

#### **Entidades Fracas (Dependentes)**  
| Entidade               | Relacionamento                     | Atributos Chave (Exemplos)                                                                 |  
|------------------------|------------------------------------|-------------------------------------------------------------------------------------------|  
| **Vaga**               | Pertence a uma `Empresa` e `Plataforma` | `id`, `cargo_padronizado`, `salario_aproximado`, `url_original`, `status_atual`, `batch_id` |  
| **ProcessoSeletivo**   | Associado a uma `Vaga`             | `id`, `etapa_atual`, `ultimo_feedback`, `data_ultima_atualizacao`                         |  
| **ScrapingLog**        | Registro de opera√ß√µes do scraper   | `id`, `url`, `timestamp`, `status`, `dados_crus` (JSON)                                   |  
| **Candidatura**        | Relaciona `Candidato` e `Vaga`     | `id`, `data_aplicacao`, `metodo_envio`, `score_compatibilidade`                           |  

---

#### **Relacionamentos Principais**  
1. **Empresa (1) ‚Üí (N) Vaga**  
   - Uma empresa pode ter m√∫ltiplas vagas.  
2. **Vaga (1) ‚Üí (N) ProcessoSeletivo**  
   - Cada vaga tem um hist√≥rico de processos.  
3. **Plataforma (1) ‚Üí (N) Vaga**  
   - Cada plataforma fonte cont√©m v√°rias vagas.  
4. **Candidato (1) ‚Üí (N) Candidatura**  
   - Voc√™ pode se candidatar a m√∫ltiplas vagas.  

---

#### **Exemplo de Entidades em SQL**  
```sql  
-- Entidade Forte: Candidato  
CREATE TABLE candidato (  
    id SERIAL PRIMARY KEY,  
    habilidades_tecnicas JSONB NOT NULL,  -- Ex: {"design_system": "avan√ßado", "figma": "experiente"}  
    pretensao_salarial NUMERIC(10,2) CHECK (pretensao_salarial > 0),  
    github_curriculo VARCHAR(255) UNIQUE  
);  

-- Entidade Fraca: Vaga  
CREATE TABLE vaga (  
    id SERIAL PRIMARY KEY,  
    cargo_padronizado VARCHAR(100) NOT NULL,  -- Ex: "PLENO|PRODUCT_DESIGNER|SENIOR|DESIGN_SYSTEM"  
    empresa_id INTEGER REFERENCES empresa(id),  
    plataforma_id INTEGER REFERENCES plataforma(id),  
    url_original VARCHAR(255) UNIQUE  
);  
```  

---

#### **4. Defini√ß√£o dos Relacionamentos**  
- **Objetivo**: Estabelecer conex√µes l√≥gicas entre entidades.  
- **Passos**:  
  - Identificar pares de entidades inter-relacionadas.  
  - Definir **cardinalidade** (1:1, 1:N, N:M).  
  - Usar **entidades associativas** para rela√ß√µes N:M (ex.: "Submiss√£o" liga "Artigo" e "Revisor").  
- **Exemplo**:  
  - "Participante" **apresenta** "Artigo" (1:N).  
  - "Artigo" **√© avaliado por** "Revisor" (N:M, via entidade "Avalia√ß√£o").  


### **4. Defini√ß√£o dos Relacionamentos para Gest√£o Inteligente de Vagas**  
*(Completo com Cardinalidades e Entidades Associativas)*  

---

#### **Relacionamentos Identificados**  

| **Entidade A**       | **Relacionamento**          | **Entidade B**          | **Cardinalidade** | **Detalhes**                                                                 |  
|----------------------|----------------------------|-------------------------|------------------|------------------------------------------------------------------------------|  
| **Candidato**        | REALIZA                    | **Candidatura**         | 1:N              | Um candidato pode ter m√∫ltiplas candidaturas.                                |  
| **Vaga**             | RECEBE                     | **Candidatura**         | 1:N              | Uma vaga pode receber m√∫ltiplas candidaturas.                                |  
| **Empresa**          | PUBLICA                    | **Vaga**                | 1:N              | Uma empresa pode publicar m√∫ltiplas vagas.                                   |  
| **Plataforma**       | HOSPEDA                    | **Vaga**                | 1:N              | Uma plataforma (LinkedIn, Glassdoor) pode listar m√∫ltiplas vagas.           |  
| **Vaga**             | POSSUI                     | **ProcessoSeletivo**    | 1:N              | Cada vaga tem um hist√≥rico de etapas do processo seletivo.                  |  
| **Vaga**             | REGISTRA                   | **ScrapingLog**         | 1:1              | Cada vaga scraped tem um log √∫nico de coleta (opcional se n√£o coletada).   |  

---

#### **Entidades Associativas para Rela√ß√µes N:M**  

1. **Compatibilidade_Candidato_Vaga** (Relacionamento impl√≠cito)  
   - **Atributos**:  
     - `score_compatibilidade` (0-100%)  
     - `data_avaliacao`  
   - **Justificativa**:  
     - Armazena o resultado do matching autom√°tico entre habilidades do candidato e requisitos da vaga.  

2. **Evento_Processo** (Para N eventos por ProcessoSeletivo)  
   - **Atributos**:  
     - `tipo` (ex: "ENTREVISTA", "FEEDBACK")  
     - `data_hora`  
     - `detalhes` (JSON com conte√∫do livre)  

---

#### **Exemplo de SQL para Relacionamentos**  

```sql  
-- Tabela associativa para Candidatura (N:M entre Candidato e Vaga)  
CREATE TABLE candidatura (  
    id SERIAL PRIMARY KEY,  
    candidato_id INTEGER NOT NULL REFERENCES candidato(id),  
    vaga_id INTEGER NOT NULL REFERENCES vaga(id),  
    data_aplicacao TIMESTAMPTZ DEFAULT NOW(),  
    score_compatibilidade DECIMAL(5,2) CHECK (score_compatibilidade BETWEEN 0 AND 100),  
    UNIQUE (candidato_id, vaga_id)  -- Evita duplicatas  
);  

-- Tabela para ProcessoSeletivo (1:N com Vaga)  
CREATE TABLE processo_seletivo (  
    id SERIAL PRIMARY KEY,  
    vaga_id INTEGER NOT NULL REFERENCES vaga(id),  
    etapa_atual VARCHAR(50) NOT NULL CHECK (etapa_atual IN ('SCREENING', 'ENTREVISTA', 'OFFER')),  
    ultimo_feedback TEXT  
);  
```  

---

#### **Diagrama Conceitual (Resumo)**  
```mermaid  
erDiagram  
    CANDIDATO ||--o{ CANDIDATURA : "REALIZA"  
    VAGA ||--o{ CANDIDATURA : "RECEBE"  
    EMPRESA ||--o{ VAGA : "PUBLICA"  
    PLATAFORMA ||--o{ VAGA : "HOSPEDA"  
    VAGA ||--o{ PROCESSO_SELETIVO : "POSSUI"  
    VAGA ||--|| SCRAPING_LOG : "REGISTRA"  
    CANDIDATURA }|--|| COMPATIBILIDADE : "AVALIA"  
```  

---


#### **5. Especifica√ß√£o dos Atributos**  
- **Objetivo**: Detalhar caracter√≠sticas das entidades.  
- **Tipos de Atributos**:  
  - **Simples**: Indivis√≠veis (ex.: "T√≠tulo da Confer√™ncia").  
  - **Compostos**: Divis√≠veis (ex.: "Endere√ßo" em rua, cidade).  
  - **Multivalorados**: M√∫ltiplos valores (ex.: "Telefones do Participante").  
  - **Chave**: Identificadores √∫nicos (ex.: "ID_Confer√™ncia").  
- **Exemplo**:  
  - Entidade "Artigo": T√≠tulo, Resumo, DataSubmiss√£o, Status.  

### **5. Especifica√ß√£o dos Atributos para Gest√£o Inteligente de Vagas**  
*(Classifica√ß√£o Completa por Entidade)*  

---

#### **1. Entidade `CANDIDATO`**  
| Atributo               | Tipo           | Descri√ß√£o                                                                 | Restri√ß√µes                          |  
|------------------------|----------------|---------------------------------------------------------------------------|-------------------------------------|  
| `id`                   | **Chave** (SERIAL) | Identificador √∫nico.                                                   | PRIMARY KEY                         |  
| `habilidades_tecnicas` | **Multivalorado** (JSONB) | Lista padronizada: `{"skill": "n√≠vel"}` (ex.: `{"figma": "avan√ßado"}`). | NOT NULL                            |  
| `pretensao_salarial`   | **Composto** (NUMERIC[2]) | Faixa m√≠nima/m√°xima: `[5000, 8000]`.                                  | CHECK (valores > 0)                 |  
| `github_curriculo`     | **Simples** (VARCHAR) | URL do reposit√≥rio com dados versionados.                              | UNIQUE                              |  
| `batch_atual`          | **Simples** (CHAR(8)) | ID do batch em execu√ß√£o (ex.: `v24s0110`).                             | Formato `vYYsWWDD`                  |  

---

#### **2. Entidade `VAGA`**  
| Atributo               | Tipo           | Descri√ß√£o                                                                 | Restri√ß√µes                          |  
|------------------------|----------------|---------------------------------------------------------------------------|-------------------------------------|  
| `id`                   | **Chave** (SERIAL) | Identificador √∫nico.                                                   | PRIMARY KEY                         |  
| `cargo_padronizado`    | **Simples** (VARCHAR) | Formato `[N√çVEL][PAPEL][SENIORIDADE][ESPECIALIDADE][CONTEXTO]`.       | Regex validation                   |  
| `salario_aproximado`   | **Composto** (JSONB) | Faixa e moeda: `{"min": 5000, "max": 8000, "moeda": "BRL"}`.          | CHECK (min < max)                   |  
| `requisitos`           | **Multivalorado** (JSONB) | `{"hard_skills": ["figma"], "soft_skills": ["comunica√ß√£o"]}`.         | NOT NULL                            |  
| `url_original`         | **Simples** (VARCHAR) | Link da vaga na plataforma fonte.                                      | UNIQUE                              |  
| `status_atual`         | **Simples** (ENUM)   | Valores: `ABERTA`, `EM_ANDAMENTO`, `FECHADA`.                          | DEFAULT 'ABERTA'                    |  

---

#### **3. Entidade `PROCESSO_SELETIVO`**  
| Atributo               | Tipo           | Descri√ß√£o                                                                 | Restri√ß√µes                          |  
|------------------------|----------------|---------------------------------------------------------------------------|-------------------------------------|  
| `id`                   | **Chave** (SERIAL) | Identificador √∫nico.                                                   | PRIMARY KEY                         |  
| `etapa_atual`          | **Simples** (ENUM) | Valores: `TRIAGEM`, `ENTREVISTA`, `TESTE_TECNICO`, `OFFER`.            | NOT NULL                            |  
| `historico_eventos`    | **Multivalorado** (JSONB[]) | Array de eventos: `[{"tipo": "feedback", "data": "2024-01-15"}]`.     | --                                  |  

---

#### **4. Entidade Associativa `CANDIDATURA`**  
| Atributo               | Tipo           | Descri√ß√£o                                                                 | Restri√ß√µes                          |  
|------------------------|----------------|---------------------------------------------------------------------------|-------------------------------------|  
| `id`                   | **Chave** (SERIAL) | Identificador √∫nico.                                                   | PRIMARY KEY                         |  
| `score_compatibilidade`| **Simples** (DECIMAL) | Porcentagem de match (ex.: `78.5`).                                   | CHECK (0 <= score <= 100)           |  
| `metodo_envio`         | **Simples** (ENUM)   | Valores: `MANUAL`, `AUTOMATICO`, `BATCH`.                              | DEFAULT 'BATCH'                     |  

---

#### **5. Tipos de Dados Especiais**  
- **JSONB**: Para atributos compostos/multivalorados (ex.: `requisitos`).  
- **ENUM**: Para campos com valores fixos (ex.: `status_atual`).  
- **Dom√≠nios Personalizados**:  
  ```sql  
  CREATE DOMAIN formato_cargo AS VARCHAR(100)  
  CHECK (VALUE ~ '^[A-Z]+\|[A-Z_]+\|[A-Z]+\|[A-Z_]+\|[A-Z_]+$');  -- Ex: "PLENO|PRODUCT_DESIGNER|SENIOR|DESIGN_SYSTEM"  
  ```  

---

#### **6. Valida√ß√£o com Regras de Neg√≥cio**  
- **Objetivo**: Garantir ader√™ncia √†s necessidades do sistema.  
- **Verificar**:  
  - Restri√ß√µes (ex.: "Nenhum artigo sem revisor").  
  - Integridade dos relacionamentos.  

### **6. Valida√ß√£o com Regras de Neg√≥cio para Gest√£o de Vagas**  
*(Verifica√ß√£o Completa de Restri√ß√µes e Integridade)*  

---

#### **1. Valida√ß√£o das Regras de Neg√≥cio**  

| **Regra de Neg√≥cio**         | **Implementa√ß√£o no Modelo**                                                                 | **Mecanismo de Valida√ß√£o**                                                                 |  
|-----------------------------|----------------------------------------------------------------------------------------|----------------------------------------------------------------------------------------|  
| **RN01**: Filtro salarial autom√°tico | Atributo `pretensao_salarial` (CANDIDATO) vs. `salario_aproximado` (VAGA)               | Trigger que marca vagas com `salario_aproximado.max < (pretensao_salarial.min * 0.8)` como "INCOMPAT√çVEL" |  
| **RN02**: Valida√ß√£o de vagas scraped | Campos obrigat√≥rios em `VAGA` (`cargo_padronizado`, `requisitos.hard_skills`)             | CHECK na tabela + valida√ß√£o no scraper (`IF NULL THEN log_error`)                      |  
| **RN03**: Preven√ß√£o de duplicatas   | Atributo `url_original` (VAGA) como UNIQUE + hash na tabela `SCRAPING_LOG`               | Constraint UNIQUE + fun√ß√£o `BEFORE INSERT` para verificar hashes existentes            |  
| **RN04**: Limite de batch           | Atributo `batch_atual` (CANDIDATO) + contagem em `CANDIDATURA`                          | Stored procedure que bloqueia novas aplica√ß√µes se `COUNT(candidaturas) >= limite_batch` |  

---

#### **2. Verifica√ß√£o de Integridade dos Relacionamentos**  

| **Relacionamento**               | **Regra de Integridade**                                                                 | **Implementa√ß√£o**                                                                 |  
|--------------------------------|------------------------------------------------------------------------------------|------------------------------------------------------------------------------|  
| **CANDIDATURA ‚Üí VAGA**         | Uma candidatura n√£o pode existir sem vaga                                          | FOREIGN KEY `vaga_id` NOT NULL + ON DELETE CASCADE                           |  
| **PROCESSO_SELETIVO ‚Üí VAGA**   | Todo processo seletivo deve estar vinculado a uma vaga ativa                       | FOREIGN KEY + CHECK (`vaga.status_atual != 'FECHADA'`)                      |  
| **SCRAPING_LOG ‚Üí VAGA**        | Logs de scraping devem referenciar vagas v√°lidas ou ser NULL (para falhas)         | FOREIGN KEY `vaga_id` ON DELETE SET NULL                                    |  

---

#### **3. Exemplos de Valida√ß√£o em SQL**  

**Trigger para RN01 (Filtro Salarial)**:  
```sql  
CREATE FUNCTION valida_salario() RETURNS TRIGGER AS $$  
BEGIN  
    IF NEW.salario_aproximado->>'max'::NUMERIC < (  
        SELECT pretensao_salarial->>'min'::NUMERIC * 0.8  
        FROM candidato  
        WHERE id = NEW.candidato_id  
    ) THEN  
        NEW.status := 'INCOMPAT√çVEL';  
    END IF;  
    RETURN NEW;  
END;  
$$ LANGUAGE plpgsql;  

CREATE TRIGGER trigger_valida_salario  
BEFORE INSERT ON candidatura  
FOR EACH ROW EXECUTE FUNCTION valida_salario();  
```  

**Constraint para RN03 (Duplicatas)**:  
```sql  
ALTER TABLE vaga  
ADD CONSTRAINT unique_url UNIQUE (url_original);  
```  

---

#### **4. Casos de Teste para Valida√ß√£o**  

| **Cen√°rio**                          | **Entrada**                                                                 | **Sa√≠da Esperada**                                                                 |  
|------------------------------------|--------------------------------------------------------------------------|--------------------------------------------------------------------------------|  
| Vaga com sal√°rio abaixo do limite   | `salario_aproximado = {"min": 3000, "max": 4000}` (Candidato: `{"min": 6000}`) | Status `INCOMPAT√çVEL` na candidatura                                           |  
| Vaga sem `hard_skills`             | `requisitos.hard_skills = NULL`                                          | Falha no INSERT (CHECK constraint)                                             |  
| URL duplicada                      | `url_original = "https://linkedin.com/jobs/123"` (j√° existente)          | Erro de constraint `unique_url`                                                |  

---

#### **7. Documenta√ß√£o e Ferramentas**  
- **Sa√≠da**:  
  - **Diagrama Entidade-Relacionamento (DER)** visualizando entidades, atributos e relacionamentos.  
  - **Ferramentas Sugeridas**:  
    - **ERwin**: Para modelos conceituais e f√≠sicos.  
    - **Lucidchart**: Colabora√ß√£o em nuvem.  
    - **Microsoft Visio**: Diagramas intuitivos.  

### **7. Documenta√ß√£o e Ferramentas para o Modelo de Gest√£o de Vagas**  

#### **1. Diagrama Entidade-Relacionamento (DER) Final**  
*(Visualiza√ß√£o Simplificada)*  

```mermaid
erDiagram
    CANDIDATO ||--o{ CANDIDATURA : "REALIZA"
    CANDIDATO {
        SERIAL id PK
        JSONB habilidades_tecnicas
        NUMERIC[] pretensao_salarial
        VARCHAR github_curriculo
        CHAR(8) batch_atual
    }
    
    VAGA ||--o{ CANDIDATURA : "RECEBE"
    VAGA {
        SERIAL id PK
        VARCHAR cargo_padronizado
        JSONB salario_aproximado
        JSONB requisitos
        VARCHAR url_original
        ENUM status_atual
    }
    
    EMPRESA ||--o{ VAGA : "PUBLICA"
    EMPRESA {
        SERIAL id PK
        VARCHAR nome
        VARCHAR setor
        VARCHAR porte
    }
    
    PLATAFORMA ||--o{ VAGA : "HOSPEDA"
    PLATAFORMA {
        SERIAL id PK
        VARCHAR nome
        VARCHAR url_base
    }
    
    VAGA ||--o{ PROCESSO_SELETIVO : "POSSUI"
    PROCESSO_SELETIVO {
        SERIAL id PK
        ENUM etapa_atual
        JSONB[] historico_eventos
    }
    
    VAGA ||--|| SCRAPING_LOG : "REGISTRA"
    SCRAPING_LOG {
        VARCHAR(64) url_hash PK
        JSONB dados_crus
        VARCHAR status
        TIMESTAMPTZ timestamp
    }
```
