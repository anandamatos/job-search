## **Processo de Modelagem de Dados: Identificação de Entidades, Atributos e Relacionamentos**  

#### **Objetivo**:  
Estruturando um modelo conceitual de banco de dados, identificando entidades, atributos e relacionamentos com base nos requisitos do sistema.  

---

### **Etapas do Processo**:  

#### **1. Definição do Minimundo**  
- **Objetivo**: Delimitar o escopo do sistema, focando nos elementos relevantes do mundo real.  
- **Ações**:  
  - Listar os principais conceitos/objetos envolvidos (ex.: "Conferência", "Participante", "Artigo").  
  - Descartar elementos irrelevantes ao sistema.


#### **Perguntas-chave para guiar a modelagem**:

1.  **Quem são os atores principais?**

    -   Candidato (que no caso sou eu o usuario do sistema, ahco que aqui importante inserir o concetido de bath ou seja, de tempos em tempos, um numero X de aplicacoes sera feitos, baseado numa versao (github) especifica dos dados que informarei para criar curriculos, cover letters, fazer analises de adequação usuario x vaga, e num futuro, a automacao disso tudo via ferramentas de automacao e AI) 
    -   Empresas Empresas que eu candidatarei
    -   Vagas que eu me aplicarei
    -   Recrutadores que serão a interface com a empresa, e nesse caso, alem dos recrutadores, gestores da vagas e outros perfis hiriequicos da empresa, inclua tambe a possibilidade de ter agentes de IA como "entrevistadores automatizados

2.  **O que deve ser rastreado sobre cada vaga?**

    -   Apenas título e salário, ou também requisitos, localização, tipo (CLT/PJ)?
    -   mesmo sabendo que o salario nunca é apresentado nas fases incias do processo, quero que tneha uma campo de salario aproximado baseado em algum app como o (glassdor, funcioarios da empresa, etc) e a faixa salarial desajada por mim
    -   cargo, ao contrario da minha primeira tabela no notion, eu quero que todas os cargos sigam algum de padrão (ex: Nivel + Papel + Senioriade + Especialisdade  + Contexto =  Product Designer Analista Sr, Design System/E-commerce)
    -   Responsabilidades e Atribuições

    -   Diferenciais / Nice to Have
    -   Soft Skills
    -   Palavras-Chave Booleanas
    -   Notas Explicativas:
    -   Beneficios

#### **Entidades Principais e Atributos**  
1. **CANDIDATO** (Você)  
   - **Dados Principais**:  
     - `hard_skills` (padronizadas: "Figma", "React", com níveis de proficiência).
     - `soft_skills` (padronizadas: "Comunicação", "LIderança", onde essas habilidades foram melhores plaicadas/desenvolvidas).  
     - `work_xp` (cargo, empresa, período, conquistas em bullets).  
     - `edu_xp` ()
     - `other_xp` ()
     - `earnings` (faixa_min, faixa_max, moeda, peso).  
     - `github_cv` (URL para JSON/YAML versionado).  
   - **Comportamento do Sistema**:  
     - Aplicações em *batch* (métricas: `batch_id` no formato `vYYsWWDD`).  
     - Interação com IA: `simulacoes_entrevista` (logs de respostas/feedback).  

2. **VAGA**  
   - **Estrutura Padronizada**:  
     - `position`: `[Nível][Papel][Senioridade][Especialidade][Contexto]` (ex: "PLENO|PRODUCT_DESIGNER|SENIOR|DESIGN_SYSTEM|E-COMMERCE").  
     - `ofers` (faixa, fonte: Glassdoor/relatos).  
     - `requirements`:  
       - `hard_skills` (lista padronizada).  
       - `soft_skills` (lista padronizada).  
       - `nice_2have` (opcional).  
     - `key_words` (ex: "DESIGN_SYSTEM AND (E-COMMERCE OR SAAS)").  
     - `benefits` (classificação: `flexibilidade`, `saude`, `bonus`).  
     - `work_model` (PRESENCIAL, HIBRIDO, REMOTO).  
     - `agreement_model` (CLT, PJ, RJU, etc.).  
   - **Rastreamento de Processo**:  
     - `step` (ex: "SCREENING", "ENTREVISTA_TECNICA", "OFFER").  
     - `events` (timestamp + status anterior + motivo).  
     - `status` (logs: mensagens RH, feedbacks, testes).  
     - `batch_id` (ex: "v23s5220" = 2023, semana 52, década 20).  

3. **EMPRESA**  
   - **Dados Estratégicos**:  
     - `sector` (TECH, FINANCEIRO, SAÚDE).  
     - `size` (STARTUP, MULTINACIONAL). 
     - `employer_numbers` (range 1-5, 6-10, 11-20...)
     - `recognition` (nota_glassdoor, premios: "GPTW", "FORBES").  
   - **Descartado**:  
     - Contatos diretos de recrutadores (foco em automação).  

4. **PROCESSO_SELETIVO** (Opcional para IA)  
   - `automated_steps` (ex: "CHATBOT_SCREENING", "TESTE_TECNICO_IA").

- **Objetivo**: Coletar necessidades dos usuários e regras de negócio.  
- **Técnicas**:  
  - **Entrevistas** com stakeholders.  
  - **Questionários** para escala ampla.  
  - **Observação direta** de processos existentes.  
- **Saída**: Documento com funcionalidades, restrições e dados necessários (ex.: "Todo artigo deve ter um revisor").  

### **Levantamento de Requisitos Atualizado para Gestão Inteligente de Vagas**  
*(Com Web Scraper e Automação Avançada)*  

---

#### **1. Requisitos Funcionais Prioritários**  
| ID  | Descrição | Detalhes Técnicos |  
|-----|-----------|-------------------|  
| **RF01** | Padronização de campos | Cargo no formato `[NÍVEL][PAPEL][SENIORIDADE][ESPECIALIDADE][CONTEXTO]` com validação Regex |  
| **RF02** | Web Scraper multi-plataforma | Extração automática de: <br>- Cargo (com conversão para padrão)<br>- Salário (Glassdoor/LinkedIn)<br>- Requisitos (Hard/Soft Skills)<br>- Benefícios |  
| **RF03** | Automação de aplicações | Batch system com: <br>- Limite configurável (ex: 25 vagas/semana)<br>- ID no formato `vYYsWWDD` |  
| **RF04** | Rastreamento inteligente | Histórico de status com:<br>- Timestamp<br>- Motivo de mudança<br>- Eventos (e-mails, feedbacks) |  

#### **2. Requisitos Não-Funcionais Críticos**  
| ID  | Descrição |  
|-----|-----------|  
| **RNF01** | Scraper resiliente com: <br>- Rotação de User-Agents<br>- Delay configurável (3-7s)<br>- Fallback para captchas |  
| **RNF02** | Versionamento de dados do candidato via GitHub (YAML/JSON) |  
| **RNF03** | Banco de dados otimizado para:<br>- Consultas booleanas complexas<br>- Armazenamento de logs brutos |  

#### **3. Regras de Negócio Atualizadas**  
| ID  | Regra | Lógica |  
|-----|-------|--------|  
| **RN01** | Filtro salarial automático | Sinalizar vagas com salário 20% abaixo da pretensão |  
| **RN02** | Validação de vagas scraped | Exigir preenchimento mínimo de:<br>- Cargo (100%)<br>- Requisitos (70%)<br>- Modalidade (100%) |  
| **RN03** | Prevenção de duplicatas | Hash de URL como chave única |

```python  
# Exemplo de pipeline completo  
def process_vaga(url):

```

---

#### **3. Identificação das Entidades**  
- **Objetivo**: Listar objetos principais que armazenarão dados.  
- **Classificação**:  
  - **Entidades Fortes**: Existem independentemente (ex.: "Conferência").  
  - **Entidades Fracas**: Dependem de outras (ex.: "Apresentação" depende de "Sessão").  
- **Exemplo**:  
  - Conferência, Participante, Sessão, Artigo, Revisor.  

### **3.1 Identificação das Entidades para Gestão Inteligente de Vagas**  
*(Com Foco em Automação e Rastreamento de Processos)*  

---

#### **Entidades Fortes (Independentes)**  
| Entidade          | Descrição                                                                 | Atributos Chave (Exemplos)                                                                 |  
|-------------------|---------------------------------------------------------------------------|-------------------------------------------------------------------------------------------|  
| **Candidato**     | Representa você (usuário do sistema)                                      | `id`, `habilidades_tecnicas`, `pretensao_salarial`, `github_curriculo`, `batch_atual`     |  
| **Empresa**       | Organizações que publicam vagas                                           | `id`, `nome`, `setor`, `porte`, `reputacao_glassdoor`                                     |  
| **Plataforma**    | Fonte das vagas (LinkedIn, Glassdoor, etc.)                               | `id`, `nome`, `url_base`, `dificuldade_scraping` (nível de complexidade para extração)     |  

#### **Entidades Fracas (Dependentes)**  
| Entidade               | Relacionamento                     | Atributos Chave (Exemplos)                                                                 |  
|------------------------|------------------------------------|-------------------------------------------------------------------------------------------|  
| **Vaga**               | Pertence a uma `Empresa` e `Plataforma` | `id`, `cargo_padronizado`, `salario_aproximado`, `url_original`, `status_atual`, `batch_id` |  
| **ProcessoSeletivo**   | Associado a uma `Vaga`             | `id`, `etapa_atual`, `ultimo_feedback`, `data_ultima_atualizacao`                         |  
| **ScrapingLog**        | Registro de operações do scraper   | `id`, `url`, `timestamp`, `status`, `dados_crus` (JSON)                                   |  
| **Candidatura**        | Relaciona `Candidato` e `Vaga`     | `id`, `data_aplicacao`, `metodo_envio`, `score_compatibilidade`                           |  

---

#### **Relacionamentos Principais**  
1. **Empresa (1) → (N) Vaga**  
   - Uma empresa pode ter múltiplas vagas.  
2. **Vaga (1) → (N) ProcessoSeletivo**  
   - Cada vaga tem um histórico de processos.  
3. **Plataforma (1) → (N) Vaga**  
   - Cada plataforma fonte contém várias vagas.  
4. **Candidato (1) → (N) Candidatura**  
   - Você pode se candidatar a múltiplas vagas.  

---

#### **Exemplo de Entidades em SQL**  
```sql  
-- Entidade Forte: Candidato  
CREATE TABLE candidato (  
    id SERIAL PRIMARY KEY,  
    habilidades_tecnicas JSONB NOT NULL,  -- Ex: {"design_system": "avançado", "figma": "experiente"}  
    pretensao_salarial NUMERIC(10,2) CHECK (pretensao_salarial > 0),  
    github_curriculo VARCHAR(255) UNIQUE  
);  

-- Entidade Fraca: Vaga  
CREATE TABLE vaga (  
    id SERIAL PRIMARY KEY,  
    cargo_padronizado VARCHAR(100) NOT NULL,  -- Ex: "PLENO|PRODUCT_DESIGNER|SENIOR|DESIGN_SYSTEM"  
    empresa_id INTEGER REFERENCES empresa(id),  
    plataforma_id INTEGER REFERENCES plataforma(id),  
    url_original VARCHAR(255) UNIQUE  
);  
```  

---

#### **4. Definição dos Relacionamentos**  
- **Objetivo**: Estabelecer conexões lógicas entre entidades.  
- **Passos**:  
  - Identificar pares de entidades inter-relacionadas.  
  - Definir **cardinalidade** (1:1, 1:N, N:M).  
  - Usar **entidades associativas** para relações N:M (ex.: "Submissão" liga "Artigo" e "Revisor").  
- **Exemplo**:  
  - "Participante" **apresenta** "Artigo" (1:N).  
  - "Artigo" **é avaliado por** "Revisor" (N:M, via entidade "Avaliação").  


### **4. Definição dos Relacionamentos para Gestão Inteligente de Vagas**  
*(Completo com Cardinalidades e Entidades Associativas)*  

---

#### **Relacionamentos Identificados**  

| **Entidade A**       | **Relacionamento**          | **Entidade B**          | **Cardinalidade** | **Detalhes**                                                                 |  
|----------------------|----------------------------|-------------------------|------------------|------------------------------------------------------------------------------|  
| **Candidato**        | REALIZA                    | **Candidatura**         | 1:N              | Um candidato pode ter múltiplas candidaturas.                                |  
| **Vaga**             | RECEBE                     | **Candidatura**         | 1:N              | Uma vaga pode receber múltiplas candidaturas.                                |  
| **Empresa**          | PUBLICA                    | **Vaga**                | 1:N              | Uma empresa pode publicar múltiplas vagas.                                   |  
| **Plataforma**       | HOSPEDA                    | **Vaga**                | 1:N              | Uma plataforma (LinkedIn, Glassdoor) pode listar múltiplas vagas.           |  
| **Vaga**             | POSSUI                     | **ProcessoSeletivo**    | 1:N              | Cada vaga tem um histórico de etapas do processo seletivo.                  |  
| **Vaga**             | REGISTRA                   | **ScrapingLog**         | 1:1              | Cada vaga scraped tem um log único de coleta (opcional se não coletada).   |  

---

#### **Entidades Associativas para Relações N:M**  

1. **Compatibilidade_Candidato_Vaga** (Relacionamento implícito)  
   - **Atributos**:  
     - `score_compatibilidade` (0-100%)  
     - `data_avaliacao`  
   - **Justificativa**:  
     - Armazena o resultado do matching automático entre habilidades do candidato e requisitos da vaga.  

2. **Evento_Processo** (Para N eventos por ProcessoSeletivo)  
   - **Atributos**:  
     - `tipo` (ex: "ENTREVISTA", "FEEDBACK")  
     - `data_hora`  
     - `detalhes` (JSON com conteúdo livre)  

---

#### **Exemplo de SQL para Relacionamentos**  

```sql  
-- Tabela associativa para Candidatura (N:M entre Candidato e Vaga)  
CREATE TABLE candidatura (  
    id SERIAL PRIMARY KEY,  
    candidato_id INTEGER NOT NULL REFERENCES candidato(id),  
    vaga_id INTEGER NOT NULL REFERENCES vaga(id),  
    data_aplicacao TIMESTAMPTZ DEFAULT NOW(),  
    score_compatibilidade DECIMAL(5,2) CHECK (score_compatibilidade BETWEEN 0 AND 100),  
    UNIQUE (candidato_id, vaga_id)  -- Evita duplicatas  
);  

-- Tabela para ProcessoSeletivo (1:N com Vaga)  
CREATE TABLE processo_seletivo (  
    id SERIAL PRIMARY KEY,  
    vaga_id INTEGER NOT NULL REFERENCES vaga(id),  
    etapa_atual VARCHAR(50) NOT NULL CHECK (etapa_atual IN ('SCREENING', 'ENTREVISTA', 'OFFER')),  
    ultimo_feedback TEXT  
);  
```  

---

#### **Diagrama Conceitual (Resumo)**  
```mermaid  
erDiagram  
    CANDIDATO ||--o{ CANDIDATURA : "REALIZA"  
    VAGA ||--o{ CANDIDATURA : "RECEBE"  
    EMPRESA ||--o{ VAGA : "PUBLICA"  
    PLATAFORMA ||--o{ VAGA : "HOSPEDA"  
    VAGA ||--o{ PROCESSO_SELETIVO : "POSSUI"  
    VAGA ||--|| SCRAPING_LOG : "REGISTRA"  
    CANDIDATURA }|--|| COMPATIBILIDADE : "AVALIA"  
```  

---


#### **5. Especificação dos Atributos**  
- **Objetivo**: Detalhar características das entidades.  
- **Tipos de Atributos**:  
  - **Simples**: Indivisíveis (ex.: "Título da Conferência").  
  - **Compostos**: Divisíveis (ex.: "Endereço" em rua, cidade).  
  - **Multivalorados**: Múltiplos valores (ex.: "Telefones do Participante").  
  - **Chave**: Identificadores únicos (ex.: "ID_Conferência").  
- **Exemplo**:  
  - Entidade "Artigo": Título, Resumo, DataSubmissão, Status.  

### **5. Especificação dos Atributos para Gestão Inteligente de Vagas**  
*(Classificação Completa por Entidade)*  

---

#### **1. Entidade `CANDIDATO`**  
| Atributo               | Tipo           | Descrição                                                                 | Restrições                          |  
|------------------------|----------------|---------------------------------------------------------------------------|-------------------------------------|  
| `id`                   | **Chave** (SERIAL) | Identificador único.                                                   | PRIMARY KEY                         |  
| `habilidades_tecnicas` | **Multivalorado** (JSONB) | Lista padronizada: `{"skill": "nível"}` (ex.: `{"figma": "avançado"}`). | NOT NULL                            |  
| `pretensao_salarial`   | **Composto** (NUMERIC[2]) | Faixa mínima/máxima: `[5000, 8000]`.                                  | CHECK (valores > 0)                 |  
| `github_curriculo`     | **Simples** (VARCHAR) | URL do repositório com dados versionados.                              | UNIQUE                              |  
| `batch_atual`          | **Simples** (CHAR(8)) | ID do batch em execução (ex.: `v24s0110`).                             | Formato `vYYsWWDD`                  |  

---

#### **2. Entidade `VAGA`**  
| Atributo               | Tipo           | Descrição                                                                 | Restrições                          |  
|------------------------|----------------|---------------------------------------------------------------------------|-------------------------------------|  
| `id`                   | **Chave** (SERIAL) | Identificador único.                                                   | PRIMARY KEY                         |  
| `cargo_padronizado`    | **Simples** (VARCHAR) | Formato `[NÍVEL][PAPEL][SENIORIDADE][ESPECIALIDADE][CONTEXTO]`.       | Regex validation                   |  
| `salario_aproximado`   | **Composto** (JSONB) | Faixa e moeda: `{"min": 5000, "max": 8000, "moeda": "BRL"}`.          | CHECK (min < max)                   |  
| `requisitos`           | **Multivalorado** (JSONB) | `{"hard_skills": ["figma"], "soft_skills": ["comunicação"]}`.         | NOT NULL                            |  
| `url_original`         | **Simples** (VARCHAR) | Link da vaga na plataforma fonte.                                      | UNIQUE                              |  
| `status_atual`         | **Simples** (ENUM)   | Valores: `ABERTA`, `EM_ANDAMENTO`, `FECHADA`.                          | DEFAULT 'ABERTA'                    |  

---

#### **3. Entidade `PROCESSO_SELETIVO`**  
| Atributo               | Tipo           | Descrição                                                                 | Restrições                          |  
|------------------------|----------------|---------------------------------------------------------------------------|-------------------------------------|  
| `id`                   | **Chave** (SERIAL) | Identificador único.                                                   | PRIMARY KEY                         |  
| `etapa_atual`          | **Simples** (ENUM) | Valores: `TRIAGEM`, `ENTREVISTA`, `TESTE_TECNICO`, `OFFER`.            | NOT NULL                            |  
| `historico_eventos`    | **Multivalorado** (JSONB[]) | Array de eventos: `[{"tipo": "feedback", "data": "2024-01-15"}]`.     | --                                  |  

---

#### **4. Entidade Associativa `CANDIDATURA`**  
| Atributo               | Tipo           | Descrição                                                                 | Restrições                          |  
|------------------------|----------------|---------------------------------------------------------------------------|-------------------------------------|  
| `id`                   | **Chave** (SERIAL) | Identificador único.                                                   | PRIMARY KEY                         |  
| `score_compatibilidade`| **Simples** (DECIMAL) | Porcentagem de match (ex.: `78.5`).                                   | CHECK (0 <= score <= 100)           |  
| `metodo_envio`         | **Simples** (ENUM)   | Valores: `MANUAL`, `AUTOMATICO`, `BATCH`.                              | DEFAULT 'BATCH'                     |  

---

#### **5. Tipos de Dados Especiais**  
- **JSONB**: Para atributos compostos/multivalorados (ex.: `requisitos`).  
- **ENUM**: Para campos com valores fixos (ex.: `status_atual`).  
- **Domínios Personalizados**:  
  ```sql  
  CREATE DOMAIN formato_cargo AS VARCHAR(100)  
  CHECK (VALUE ~ '^[A-Z]+\|[A-Z_]+\|[A-Z]+\|[A-Z_]+\|[A-Z_]+$');  -- Ex: "PLENO|PRODUCT_DESIGNER|SENIOR|DESIGN_SYSTEM"  
  ```  

---

#### **6. Validação com Regras de Negócio**  
- **Objetivo**: Garantir aderência às necessidades do sistema.  
- **Verificar**:  
  - Restrições (ex.: "Nenhum artigo sem revisor").  
  - Integridade dos relacionamentos.  

### **6. Validação com Regras de Negócio para Gestão de Vagas**  
*(Verificação Completa de Restrições e Integridade)*  

---

#### **1. Validação das Regras de Negócio**  

| **Regra de Negócio**         | **Implementação no Modelo**                                                                 | **Mecanismo de Validação**                                                                 |  
|-----------------------------|----------------------------------------------------------------------------------------|----------------------------------------------------------------------------------------|  
| **RN01**: Filtro salarial automático | Atributo `pretensao_salarial` (CANDIDATO) vs. `salario_aproximado` (VAGA)               | Trigger que marca vagas com `salario_aproximado.max < (pretensao_salarial.min * 0.8)` como "INCOMPATÍVEL" |  
| **RN02**: Validação de vagas scraped | Campos obrigatórios em `VAGA` (`cargo_padronizado`, `requisitos.hard_skills`)             | CHECK na tabela + validação no scraper (`IF NULL THEN log_error`)                      |  
| **RN03**: Prevenção de duplicatas   | Atributo `url_original` (VAGA) como UNIQUE + hash na tabela `SCRAPING_LOG`               | Constraint UNIQUE + função `BEFORE INSERT` para verificar hashes existentes            |  
| **RN04**: Limite de batch           | Atributo `batch_atual` (CANDIDATO) + contagem em `CANDIDATURA`                          | Stored procedure que bloqueia novas aplicações se `COUNT(candidaturas) >= limite_batch` |  

---

#### **2. Verificação de Integridade dos Relacionamentos**  

| **Relacionamento**               | **Regra de Integridade**                                                                 | **Implementação**                                                                 |  
|--------------------------------|------------------------------------------------------------------------------------|------------------------------------------------------------------------------|  
| **CANDIDATURA → VAGA**         | Uma candidatura não pode existir sem vaga                                          | FOREIGN KEY `vaga_id` NOT NULL + ON DELETE CASCADE                           |  
| **PROCESSO_SELETIVO → VAGA**   | Todo processo seletivo deve estar vinculado a uma vaga ativa                       | FOREIGN KEY + CHECK (`vaga.status_atual != 'FECHADA'`)                      |  
| **SCRAPING_LOG → VAGA**        | Logs de scraping devem referenciar vagas válidas ou ser NULL (para falhas)         | FOREIGN KEY `vaga_id` ON DELETE SET NULL                                    |  

---

#### **3. Exemplos de Validação em SQL**  

**Trigger para RN01 (Filtro Salarial)**:  
```sql  
CREATE FUNCTION valida_salario() RETURNS TRIGGER AS $$  
BEGIN  
    IF NEW.salario_aproximado->>'max'::NUMERIC < (  
        SELECT pretensao_salarial->>'min'::NUMERIC * 0.8  
        FROM candidato  
        WHERE id = NEW.candidato_id  
    ) THEN  
        NEW.status := 'INCOMPATÍVEL';  
    END IF;  
    RETURN NEW;  
END;  
$$ LANGUAGE plpgsql;  

CREATE TRIGGER trigger_valida_salario  
BEFORE INSERT ON candidatura  
FOR EACH ROW EXECUTE FUNCTION valida_salario();  
```  

**Constraint para RN03 (Duplicatas)**:  
```sql  
ALTER TABLE vaga  
ADD CONSTRAINT unique_url UNIQUE (url_original);  
```  

---

#### **4. Casos de Teste para Validação**  

| **Cenário**                          | **Entrada**                                                                 | **Saída Esperada**                                                                 |  
|------------------------------------|--------------------------------------------------------------------------|--------------------------------------------------------------------------------|  
| Vaga com salário abaixo do limite   | `salario_aproximado = {"min": 3000, "max": 4000}` (Candidato: `{"min": 6000}`) | Status `INCOMPATÍVEL` na candidatura                                           |  
| Vaga sem `hard_skills`             | `requisitos.hard_skills = NULL`                                          | Falha no INSERT (CHECK constraint)                                             |  
| URL duplicada                      | `url_original = "https://linkedin.com/jobs/123"` (já existente)          | Erro de constraint `unique_url`                                                |  

---

#### **7. Documentação e Ferramentas**  
- **Saída**:  
  - **Diagrama Entidade-Relacionamento (DER)** visualizando entidades, atributos e relacionamentos.  
  - **Ferramentas Sugeridas**:  
    - **ERwin**: Para modelos conceituais e físicos.  
    - **Lucidchart**: Colaboração em nuvem.  
    - **Microsoft Visio**: Diagramas intuitivos.  

### **7. Documentação e Ferramentas para o Modelo de Gestão de Vagas**  

#### **1. Diagrama Entidade-Relacionamento (DER) Final**  
*(Visualização Simplificada)*  

```mermaid
erDiagram
    CANDIDATO ||--o{ CANDIDATURA : "REALIZA"
    CANDIDATO {
        SERIAL id PK
        JSONB habilidades_tecnicas
        NUMERIC[] pretensao_salarial
        VARCHAR github_curriculo
        CHAR(8) batch_atual
    }
    
    VAGA ||--o{ CANDIDATURA : "RECEBE"
    VAGA {
        SERIAL id PK
        VARCHAR cargo_padronizado
        JSONB salario_aproximado
        JSONB requisitos
        VARCHAR url_original
        ENUM status_atual
    }
    
    EMPRESA ||--o{ VAGA : "PUBLICA"
    EMPRESA {
        SERIAL id PK
        VARCHAR nome
        VARCHAR setor
        VARCHAR porte
    }
    
    PLATAFORMA ||--o{ VAGA : "HOSPEDA"
    PLATAFORMA {
        SERIAL id PK
        VARCHAR nome
        VARCHAR url_base
    }
    
    VAGA ||--o{ PROCESSO_SELETIVO : "POSSUI"
    PROCESSO_SELETIVO {
        SERIAL id PK
        ENUM etapa_atual
        JSONB[] historico_eventos
    }
    
    VAGA ||--|| SCRAPING_LOG : "REGISTRA"
    SCRAPING_LOG {
        VARCHAR(64) url_hash PK
        JSONB dados_crus
        VARCHAR status
        TIMESTAMPTZ timestamp
    }
```
